```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import re
import pathlib
pathlib.Path().resolve()
ranks = [4,6,8,10,12]
scale = [4096, 8192,16384, 32768, 65536, 131072, 262144, 524288]
threads = [2,4,8]
idx_max = 1
thread = [1,2,4,8]
```

```python
#extract ranks, threads and time out of ssca1 output file
def get_rank_thread_time(file_path):
    input_lines = []
    rank_num = 0
    elapsed_time = []
    thread_num = 0
    with open(file_path,'r') as result_file:
        input_lines = result_file.readlines()
    for line in input_lines:
        match_rank = re.search('world size is ([0-9]*)', line)
        match_time = re.search('time ([0-9.;]*)', line)
        if match_rank:
            rank_num = match_rank.group(1)
        if match_time:
            words = line.split()
            elapsed_time = float(match_time.group(1))
    return rank_num, thread_num, elapsed_time

```

```python
class eval_collect:
    def __init__(self, filename,results_path ='../results/', thread=''):
        masterlist =[]
        for j in scale:
                masterlist.append(get_rank_thread_time(results_path+filename+str(j)+' idx'+str(1)+thread+'.txt'))
        self.rank_num,self.thread_num, self.elapsed_ms = zip(*masterlist)
    
```

```python
#plot
def plot_no_show(x,y,c='red',cc='r--',label_=''):
    x = [float(i) for i in x]
    plt.scatter(x,y,c=c)
    plt.plot(x,y,cc,label=label_)
    
def plot_3(x1,y1,x2,y2,x3,y3, title ='', label1='', label2='', label3='', xlabel_='Skalierung'):
    plot_no_show(x1,y1,c='red',label_=label1)
    plot_no_show(x2,y2,c='green',cc='g--',label_=label2)
    plot_no_show(x3,y3,c='blue',cc='b--',label_=label3)
    plt.ylabel('Zeit [ms]')
    plt.xlabel(xlabel_)
    plt.title(title)
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.grid()
    scale_local= scale
    del scale_local[::2]
    plt.gca().set_xticks(scale_local)
    plt.gca().set_xticklabels(scale_local)
    plt.show()


```

```python
scale=[ 16, 32, 64, 128, 256, 512, 1024,2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144, 524288]
gaspi_ring_2_thread = eval_collect('scale_gaspi_mfile_ib0', thread='2')
mpi_ring_2_thread = eval_collect('scale_mpi_mfile_ib0', thread='2')
shmem_ring_2_thread = eval_collect('scale_shmem_mfile_ib0', thread='2')
plot_3(scale,gaspi_ring_2_thread.elapsed_ms,\
       scale,mpi_ring_2_thread.elapsed_ms,\
       scale,shmem_ring_2_thread.elapsed_ms,\
       '','GASPI','MPI',"OpenSHMEM")
```
