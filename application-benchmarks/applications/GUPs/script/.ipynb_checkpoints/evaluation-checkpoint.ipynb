```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import re
import pathlib
pathlib.Path().resolve()
ranks = [2,3,4,5,6]
threads = [1,2,4,8,16]
idx_max = 3
scale = [1,10,100,1000,10000,100000, 1000000]
thread = [1,2,4,8]
```

```python
#extract ranks, threads and time out of ssca1 output file
def get_rank_thread_time(file_path):
    input_lines = []
    rank_num = 0
    elapsed_time = []
    thread_num = 0
    with open(file_path,'r') as result_file:
        input_lines = result_file.readlines()
    for line in input_lines:
        match_rank = re.search('world size is ([0-9]*)', line)
        match_time = re.search('([0-9.;]*) Billion', line)
        if match_rank:
            rank_num = match_rank.group(1)
        if match_time:
            words = line.split()
            elapsed_time = float(match_time.group(1))
            break
    thread_num = 0
    return rank_num, thread_num, elapsed_time



def get_rank_thread_time_weak(file_path):
    input_lines = []
    elapsed_time = []
    scale_num = 0
    with open(file_path,'r') as result_file:
        input_lines = result_file.readlines()
    for line in input_lines:
        match_scale = re.search('Scale Size: ([0-9]*)', line)
        match_time = re.search('([0-9.;]*) seconds', line)
        if match_scale:
            scale_num = match_scale.group(1)
        if match_time:
            words = line.split()
            elapsed_time = float(match_time.group(1))
            break
    bandwith = float(scale_num)/elapsed_time
    return scale_num, elapsed_time, bandwith
```

```python
class eval_collect:
    def __init__(self, filename,results_path ='../results/', thread='',rank_total=ranks):
        masterlist =[]
        for j in rank_total:
            for i in range(1, idx_max):
                masterlist.append(get_rank_thread_time(results_path+filename+str(j)+thread+' idx'+str(i)+'.txt'))
        self.rank_num,self.thread_num, self.elapsed_ms = zip(*masterlist)
        self.elapsed_ms_mean = [0] * len(rank_total)
        for idx,i in enumerate(rank_total):
            for jdx,rank in enumerate(self.rank_num):
                if str(rank) == str(i):
                    self.elapsed_ms_mean[idx] += self.elapsed_ms[jdx]
            self.elapsed_ms_mean[idx]=self.elapsed_ms_mean[idx]/idx_max 

class eval_collect_weak:
    def __init__(self,filename,results_path ='../results/', help=''):
        masterlist =[]
        for j in scale:
            for i in range(1, idx_max):
                masterlist.append(get_rank_thread_time_weak(results_path+filename+'2'+str(j)+' idx'+str(i)+'.txt'))
        self.scale_num, self.elapsed_ms, self.bandwith = zip(*masterlist)
        self.elapsed_ms_mean = [0] * len(scale)
        self.bandwith_mean = [0] * len(scale)
        for idx,i in enumerate(scale):
            for jdx,sc in enumerate(self.scale_num):
                if str(sc) == str(i):
                    self.bandwith_mean[idx] += self.bandwith[jdx]
                    self.elapsed_ms_mean[idx] += self.elapsed_ms[jdx]
            self.elapsed_ms_mean[idx]=self.elapsed_ms_mean[idx]/idx_max 
            self.bandwith_mean[idx]=self.bandwith_mean[idx]/idx_max 
```

```python
#plot
def plot_no_show(x,y,c='red',cc='r--',label_=''):
    x = [float(i) for i in x]
    plt.scatter(x,y,c=c)
    plt.plot(x,y,cc,label=label_)
    
def plot_1(x1,y1, title ='', label1='', xlabel_='Prozesse'):
    plot_no_show(x1,y1,c='red',label_=label1)
    plt.ylabel('GUP/s')
    plt.xlabel(xlabel_)
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
    
def plot_2(x1,y1,x2,y2, title ='', label1='', label2='', xlabel_='Prozesse'):
    plot_no_show(x1,y1,c='red',label_=label1)
    plot_no_show(x2,y2,c='green',cc='g--',label_=label2)
    plt.ylabel('GUP/s')
    plt.xlabel(xlabel_)
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
    
    
def plot_3(x1,y1,x2,y2,x3,y3, title ='', label1='', label2='', label3='', xlabel_ ='Prozesse',ylabel_='GUP/s',mylog=False):
    plot_no_show(x1,y1,c='red',label_=label1)
    plot_no_show(x2,y2,c='green',cc='g--',label_=label2)
    plot_no_show(x3,y3,c='blue',cc='b--',label_=label3)
    plt.ylabel(ylabel_)
    plt.xlabel(xlabel_)
    plt.title(title)
    plt.legend()
    plt.grid()
    if mylog:
        plt.xscale('log')
        plt.yscale('log')
    plt.show()
    
def plot_4(x1,y1,x2,y2,x3,y3,x4,y4, title ='',\
           label1='', label2='', label3='',label4=''):
    plot_no_show(x1,y1,c='red',label_=label1)
    plot_no_show(x2,y2,c='green',cc='g--',label_=label2)
    plot_no_show(x3,y3,c='blue',cc='b--',label_=label3)
    plot_no_show(x4,y4,c='yellow',cc='y--',label_=label4) 
    plt.ylabel('GUP/s')
    plt.xlabel('Prozesse')
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

def plot_6(x1,y1,x2,y2,x3,y3,x4,y4,x5,y5,x6,y6, title ='',\
           label1='', label2='', label3='',label4='', label5='', label6='', xlabel_ ='Prozesse'):
    plot_no_show(x1,y1,c='red',label_=label1)
    plot_no_show(x2,y2,c='green',cc='g--',label_=label2)
    plot_no_show(x3,y3,c='blue',cc='b--',label_=label3)
    plot_no_show(x4,y4,c='yellow',cc='y--',label_=label4) 
    plot_no_show(x5,y5,c='magenta',cc='m--',label_=label5)
    plot_no_show(x6,y6,c='cyan',cc='c--',label_=label6) 
    plt.xlabel(xlabel_)
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
```
